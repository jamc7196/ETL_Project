```python
# Import dependencies
import pandas as pd
```

```python
# Create a DataFrame from CSV file and display it
match_df = pd.DataFrame(pd.read_csv('WorldCupMatches.csv', encoding = 'utf8'))
match_df.head()
```

```python
# Replace Federal Republic of Germany to Germany
match_df = match_df.replace({'Home Team Initials': {'FRG': 'GER'},
                             'Away Team Initials': {'FRG': 'GER'},
                             'Home Team Name': {'Germany FR': 'Germany'},
                             'Away Team Name': {'Germany FR': 'Germany'}})


match_df.head()
```

```python
# Replace Federal Republic of Germany to Germany
cup_df = cup_df.replace({'Winner': {'Germany FR': 'Germany'},
                         'Runners-Up': {'Germany FR': 'Germany'},
                         'Third': {'Germany FR': 'Germany'},
                         'Fourth': {'Germany FR': 'Germany'}})
cup_df
```

## Referee Table

```python
# Create DataFrames to store the MatchID and the Referee as Name
referees = match_df[['MatchID', 'Referee']]
referees = referees.rename(columns = {'Referee':'Name'})

ast1 = match_df[['MatchID', 'Assistant 1']]
ast1 = ast1.rename(columns = {'Assistant 1':'Name'})

ast2 = match_df[['MatchID', 'Assistant 2']]
ast2 = ast2.rename(columns = {'Assistant 2':'Name'})
```

```python
# Add a column to each DataFrame stating the referee rank
referees.insert(loc=1, column='Status', value=['Referee' for i in range(referees.shape[0])])
ast1.insert(loc=1, column='Status', value=['Assistant 1' for i in range(referees.shape[0])])
ast2.insert(loc=1, column='Status', value=['Assistant 2' for i in range(referees.shape[0])])
```

```python
# Drop empty rows
referees = referees.dropna()
ast1 = ast1.dropna()
ast2 = ast2.dropna()
```

```python
# Add the Assistant 1 DataFrame to the Referee
referees = referees.append(ast1, ignore_index=True)
```

```python
# Add the Assistant 2 DataFrame to the Referee
referees = referees.append(ast2, ignore_index=True)
```

```python
# Set Column Datatype
referees['MatchID'] = referees['MatchID'].astype(int)
```

```python
# Check the Referee table
referees.head()
```

```python
# Check the Referee table
referees.groupby(['MatchID'])['Name'].unique()
```

```python
referees = referees.drop_duplicates(keep = 'first')
```

```python
referees
```

```python
# Save to a CSV file without index
referees.to_csv('Tables/referees.csv', index = False)
```

## Participants

```python
# del participants
# del participants_1
# del participants_2
# del dictionary
# del dictionary1
# del dictionary2
# del participants_final
# del winners
# del second
# del third
# del fourth
```

```python
# Create DataFrames to store the Year and the semi-finalists as country_id
winners = cup_df[['Year', 'Winner']]
winners = winners.rename(columns = {'Winner':'Country'})

second = cup_df[['Year', 'Runners-Up']]
second = second.rename(columns = {'Runners-Up':'Country'})

third = cup_df[['Year', 'Third']]
third = third.rename(columns = {'Third':'Country'})

fourth = cup_df[['Year', 'Fourth']]
fourth = fourth.rename(columns = {'Fourth':'Country'})
```

```python
# Add a column to each DataFrame stating the semi-finalists' Place
winners.insert(loc=2, column='Place', value=['Winner' for i in range(winners.shape[0])])
second.insert(loc=2, column='Place', value=['Second' for i in range(second.shape[0])])
third.insert(loc=2, column='Place', value=['Third' for i in range(third.shape[0])])
fourth.insert(loc=2, column='Place', value=['Fourth' for i in range(fourth.shape[0])])
```

```python
# Drop empty rows
winners = winners.dropna()
second = second.dropna()
third = third.dropna()
fourth = fourth.dropna()
```

```python
# Append all the DataFrames to keep as one
participants_1 = winners.append(second, ignore_index=True)
participants_2 = participants_1.append(third, ignore_index = True)
participants = participants_2.append(fourth, ignore_index = True)
```

```python
# Sorth the DataFrame's values by year
participants = participants.sort_values(by='Year')
# participants
```

```python
# Create a dictionary dataframe with every world cup participant and its initials
dictionary1 = match_df[['Year','Home Team Name','Home Team Initials']]
dictionary1 = dictionary1.dropna()
dictionary1 = dictionary1.drop_duplicates(subset=['Year', 'Home Team Name'], keep='first')

# Format names
dictionary1['Home Team Name'] = dictionary1['Home Team Name'].replace({'rn">Bosnia and Herzegovina': 'Bosnia and Herzegovina',
                                                                          'rn">Republic of Ireland': 'Republic of Ireland',
                                                                          'rn">Serbia and Montenegro': 'Serbia and Montenegro',
                                                                          'rn">Trinidad and Tobago': 'Trinidad and Tobago',
                                                                          'rn">United Arab Emirates': 'United Arab Emirates'})

# Rename the DataFrames columns to match the participant's
dictionary1 = dictionary1.rename(columns = {'Home Team Name':'Country',
                                          'Home Team Initials':'country_id'})

# Format year
dictionary1['Year'] = dictionary1['Year'].astype(int)

# Display the dataframe
dictionary1
```

```python
# Create a dictionary dataframe with every world cup participant and its initials
dictionary2 = match_df[['Year','Away Team Name','Away Team Initials']]
dictionary2 = dictionary2.dropna()
dictionary2 = dictionary2.drop_duplicates(subset=['Year', 'Away Team Name'], keep='first')

# Format names
dictionary2['Away Team Name'] = dictionary2['Away Team Name'].replace({'rn">Bosnia and Herzegovina': 'Bosnia and Herzegovina',
                                                                          'rn">Republic of Ireland': 'Republic of Ireland',
                                                                          'rn">Serbia and Montenegro': 'Serbia and Montenegro',
                                                                          'rn">Trinidad and Tobago': 'Trinidad and Tobago',
                                                                          'rn">United Arab Emirates': 'United Arab Emirates'})

# Rename the DataFrames columns to match the participant's
dictionary2 = dictionary2.rename(columns = {'Away Team Name':'Country',
                                          'Away Team Initials':'country_id'})

# Format year
dictionary2['Year'] = dictionary2['Year'].astype(int)

# Display the dataframe
dictionary2
```

```python
# Append DataFrames to keep as one
dictionary = dictionary1.append(dictionary2, ignore_index = True)

# Drop empty-valued cells
dictionary = dictionary.dropna()

# Drop duplicates
dictionary = dictionary.drop_duplicates(subset=['Year', 'Country'], keep='first')

# Display DataFrame
dictionary
```

```python
# Create a dictionary with key as country name and value as initials
initials = dictionary.set_index('Country')['country_id'].to_dict()
# initials
```

```python
# Create a list of initials for each world cup semi-finalist
name = participants['Country']
country_initials = []
for i in name:
    country_initials.append(initials[i]) 
#     print(f'{i}, {initials[i]}')

# print(len(country_initials))
```

```python
# Add the semi-finalists list to the participants DataFrame
participants['country_id'] = country_initials
```

```python
# Display the dataframe
participants
```

```python
# Add a column to each DataFrame stating the semi-finalists' Place
dictionary.insert(loc=2, column='Place', value=['Non-Finalist' for i in range(dictionary.shape[0])])
```

```python
# Drop empty rows
dictionary = dictionary.dropna()
```

```python
# Append the DataFrame into participants
participants = participants.append(dictionary, ignore_index = True)
# participants
```

```python
# Show data for comparison
participants.groupby(['Year'])['Country'].value_counts()
```

```python
# Show data for comparison
participants.groupby(['Year'])['Place'].value_counts()
```

## Events

```python
# del event
# del events
# del event_df
```

```python
goals_df = player_df.dropna().copy()
```

```python
goals_df.head()
```

```python
events = goals_df[['Event', 'MatchID', 'Team Initials', 'Player Name']]

events[['Event1', 'Event2', 'Event3', 'Event4']] = events['Event'].str.split(expand = True)

events
```

```python
ev1 = events[['MatchID', 'Team Initials', 'Player Name', 'Event1']].dropna().rename(columns = {'Event1':'Event'})

ev2 = events[['MatchID', 'Team Initials', 'Player Name', 'Event2']].dropna().rename(columns = {'Event2':'Event'})

ev3 = events[['MatchID', 'Team Initials', 'Player Name', 'Event3']].dropna().rename(columns = {'Event3':'Event'})

ev4 = events[['MatchID', 'Team Initials', 'Player Name', 'Event4']].dropna().rename(columns = {'Event4':'Event'})
```

```python
events_df = pd.concat([ev1, ev2, ev3, ev4], axis = 0)
events_df
```

```python
events_df = events_df[['Event', 'MatchID', 'Team Initials', 'Player Name']]
events_df = events_df.rename(columns = {'Player Name':'Name', 'Team Initials':'country_id'})
events_df
```

```python
events_df = events_df.merge(player_name, on='Name')
```

```python
events_df = events_df.drop_duplicates(subset = ['Event', 'MatchID', 'Name'], keep = 'first')
events_df
```

```python
events_df = events_df[['Event', 'MatchID', 'country_id', 'player_id']]
events_df
```

```python
events_df = events_df.reset_index().drop(columns = ['index'])
events_df
```

```python
events_df = events_df.reset_index().rename(columns={'index':'event_id'})
events_df
```

```python
# Save to a CSV file without index
events_df.to_csv('Tables/events.csv', index = False)
```
